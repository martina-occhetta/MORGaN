```python
from pathlib import Path

import pandas as pd
import numpy as np
import mygene
```

### Results

```python
results_dir = "../results/alzheimers/CPDB/seed_0/iter_1"
```

```python
y_true = np.load(Path(results_dir, "y_true.npy"))
y_pred = np.load(Path(results_dir, "y_pred.npy"))
y_proba = np.load(Path(results_dir, "y_proba.npy"))
gene_names = np.load(Path(results_dir, "test_gene_names.npy"))
```

```python
predictions = pd.DataFrame({
    "gene_name": gene_names,
    "y_true": y_true,
    "y_pred": y_pred,
    "y_proba": y_proba
})
```

```python
predictions.sort_values(by="y_proba", ascending=False, inplace=True)
```

```python
predictions[predictions["y_true"] == 0].head(25)
```

```python
mg = mygene.MyGeneInfo()
gene_info = {}

for gene in predictions[predictions["y_true"] == 0].head(25)["gene_name"]:
    result = mg.query(
        gene, species="human", fields="symbol,name,summary,go,uniprot", size=1
    )
    if result["hits"]:
        gene_info[gene] = result["hits"][0]
    else:
        gene_info[gene] = "No information found"

print("\nBiological information for each gene:")
for gene, info in gene_info.items():
    print(f"\nGene: {gene}")
    if isinstance(info, dict):
        print("Symbol:", info.get("symbol", "N/A"))
        print("Name:", info.get("name", "N/A"))
        print("Summary:", info.get("summary", "N/A"))
        if "go" in info:
            print("GO Annotations:")
            for category, terms in info["go"].items():
                print(f"  {category.upper()}:")
                # 'terms' may be a list or a single dict
                if isinstance(terms, list):
                    for term in terms:
                        print("    -", term.get("term", "N/A"))
                elif isinstance(terms, dict):
                    print("    -", terms.get("term", "N/A"))
        if "uniprot" in info:
            print("UniProt:", info["uniprot"])
    else:
        print(info)
```
